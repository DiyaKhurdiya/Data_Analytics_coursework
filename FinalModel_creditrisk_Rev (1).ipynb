```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data
```

```python

X = data.iloc[: , 3:13].values
X1 = data.iloc[: , 3:13].values

y = data.iloc[:, 13].values
y1 = data.iloc[:, 13].values
X,y
```

```python
X[:,2], X1[:,2]
```

```python
lb = LabelEncoder()
X[:,2] = lb.fit_transform(X[:,2])
X[:,1] = lb.fit_transform(X[:,1])

X1[:,2] = lb.fit_transform(X1[:,2])
X1[:,1] = lb.fit_transform(X1[:,1])
```

```python
from sklearn.compose import ColumnTransformer
#LabelRncoder = 
#OneHotEncoder 
ct = ColumnTransformer([("Country", OneHotEncoder(), [1])], remainder = 'passthrough')
X = ct.fit_transform(X)
X1 = ct.fit_transform(X1)
X 
```

```python
#Multi-collinearity 
#n sub categories, you need n-1 categories
```

```python
#X = pd.DataFrame(X, columns = ["X2", "X3", "Credit Score", "Gender", "Age","Tenure", "Balance", "NumofProducts", "HasCrCard", "IsActive Member", "Estimated Salary"])
X1 = pd.DataFrame(X1, columns = ["X1","X2", "X3", "Credit Score", "Gender", "Age","Tenure", "Balance", "NumofProducts", "HasCrCard", "IsActive Member", "Estimated Salary"])

X1
```

```python
X1 = X1.drop("X1", axis = 1)
X1
#Apple = 1- Bananas - Ornage
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import statsmodels.api as sm
X_train = sm.add_constant(X_train)
#add_constant add a column with 1 values
X_test = sm.add_constant(X_test)
classifier = sm.Logit(y_train, X_train) #Logit
model = classifier.fit()
model.summary()
```

```python
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)
y_pred = (y_pred > 0.45) #
print("Accuracy", metrics.accuracy_score(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt ='')
```

Decision Tree classifer

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size = 0.2, random_state = 0)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train1 = sc.fit_transform(X_train1)
X_test1 = sc.transform(X_test1)
```

```python
clf  = DecisionTreeClassifier(max_depth = 6)
clf = clf.fit(X_train1, y_train1)
Y_pred = clf.predict(X_test1)
```

```python
print("Accuracy", metrics.accuracy_score(y_test1, Y_pred))
```

```python
new_customer = clf.predict(sc.transform(np.array([[0,0,1,600,1,40,3,60000,2,1,1,50000]])))
new_customer
```
