#Multi-linear regression between the independent variable (SpecialProjectsCount), control variable (EngagementSurvey) and dependent variable (Salary).

```python
#Residual Plot
plt.scatter(data2['SpecialProjectsCount'], data2['Salary'])
plt.axhline(y=0.0, color='r', linestyle='-')
plt.show()
```

```python
X = data2[['SpecialProjectsCount', 'EngagementSurvey']]
```

```python
y = data2['Salary']
```

```python
X = sm.add_constant(X)
#Those who made python forgot to add the constant that is Y= mX+c
model = sm.OLS(y, X)
results = model.fit()
results.params
```

```python
results.summary()
```

```python
sns.heatmap(data2[['SpecialProjectsCount','EngagementSurvey', 'Salary','PerfScoreID']].corr(), annot=True, fmt='.2g')
plt.tight_layout()
```

```python
#Prediction based on model
predicted_salary = results.predict(X)
print(predicted_salary)
```

```python
#Computation of error
data2['pred'] = predicted_salary
data2['error'] = y - predicted_salary
data2
```

```python
#Compute Mean Square
import sklearn.metrics
sklearn.metrics.mean_squared_error(y, data2['pred'])
```

```python
#Residual Plot
plt.scatter(data2['pred'], data2['error'])
plt.axhline(y=0.0, color='r', linestyle='-')
plt.show()
```

Since we have discrete values for SpecialProjCount we do not see a curve and instead parallel lines drawn. The residual plot does not show any relationship, and since R-suared values is 26%, we can say that there is some positive relationship between salary and special projects count.
