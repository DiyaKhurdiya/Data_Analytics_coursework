```python
df = df_1.drop('Date', axis =1)
df2 = df.drop('NIFTY', axis= 1)
df2
```

```python
X = df2
X.head()
```

```python
Y = df['NIFTY']
Y
```

```python
import statsmodels.api as sm
#This adds the constant term beta0 to the Multiple Linear Regression
X = sm.add_constant(X)
model = sm.OLS(Y,X).fit()
model.summary()
```

```python
from statsmodels.multivariate.pca import PCA
pc = PCA(df, #it should only have X's
         ncomp = 2,
         standardize = False #cuz the data has different units Rs./number etc.
)
```

```python
df_comp = pc.loadings.T # pc.loadings are the eigen vectors tha we computed
#We are transposing it so that we can see it nicely
df_comp
```

```python
#PCA Scores
X_scores = pc.factors
X_scores
```

```python
sns.heatmap(X_scores.corr())
plt.show() #There is no correlation because the PCs are orthogonal to each other and therefore multicollinearity disappears.
```

```python
#First run the PCA and then do the following lines of code
# This adds the constant term beta0 to the Mjultiple Linear Regression
X_pca=sm.add_constant(X_scores)
model = sm.OLS(Y,X_pca).fit()
model.summary()
```

N_comp = 2 is the best optimal value for n since, the confidence interval for comp_0 and comp_1 does not include 0

```python
print('Accuracy', model.rsquared*100)
```
